```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import warnings

warnings.filterwarnings("ignore")
```

## Clean Data

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
cancer
```

```python
len(cancer["target"])
```

```python
# 1. create the dataframe
df_cancer = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
df_cancer
```

```python
df_cancer.describe()
```

## Model building 

```python
def build_model(model, df):
    X, y = df, cancer["target"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=46)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    fpr = fp / (fp + tn)
    print("Accuracy: {:.4f}".format(accuracy_score(y_test, y_pred)))
    print("Precision: {:.4f}".format(precision_score(y_test, y_pred)))
    print("Recall: {:.4f}".format(recall_score(y_test, y_pred)))
    print("F1-score: {:.4f}".format(f1_score(y_test, y_pred)))
    print("AUC: {:.4f}".format(roc_auc_score(y_test, y_pred)))
    print("False Positive Rate: {:.4f}".format(fpr))
    print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
    print("Classification Report: \n", classification_report(y_test, y_pred))

```

```python
model = SVC()
build_model(model, df_cancer)
```

### Dimension reduction with PCA

```python
#  4.1 PCA
pca = PCA(n_components=2)
pca.fit(df_cancer)
pca_feats = pca.transform(df_cancer)
pca_feats
```

```python
# 4.2 select the feature
model = SVC()
build_model(model, pca_feats)
```

## Hyperparameters

```python
from sklearn.model_selection import GridSearchCV, KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
params = {
    "kernel": ["rbf", "linear"],
    "gamma": [1e-3, 1e-4],
    "C":[1, 10, 100]
}
svc = SVC()
gs = GridSearchCV(svc, params, n_jobs=-1, verbose=2, cv=kf)
gs.fit(pca_feats, cancer["target"])
print(gs.best_params_, gs.best_score_)
```
